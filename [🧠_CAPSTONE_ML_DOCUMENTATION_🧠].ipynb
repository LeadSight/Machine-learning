### **2.4.2 Analisis Kualitas Data Lanjutan (Kategorikal)**

**ðŸ’¡ Rationale**

Tahap ini berfokus pada analisis *missing value* tersembunyi (*implicit*) yang teridentifikasi pada Tahap 2.2. Meskipun tidak ada nilai `NaN` eksplisit, banyak *feature* kategorikal menggunakan label **"unknown"** atau **"nonexistent"** yang persentasenya wajib dihitung. Kuantifikasi masalah ini menentukan strategi penanganan data oleh rekan tim di **Tahap 3: Data Preparation**.

````python
# Kolom kategorikal yang diketahui mengandung nilai 'unknown'
unknown_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact']

print("--- Persentase Missing Value 'unknown' per Kolom ---")
print("----------------------------------------------------")

# 1. Loop melalui kolom umum yang mengandung 'unknown'
for col in unknown_cols:
    # Menggunakan df secara konsisten
    unknown_count = df[df[col] == 'unknown'].shape[0]
    total_rows = df.shape[0]
    percentage = (unknown_count / total_rows) * 100

    if percentage > 0:
        print(f"'{col}': {unknown_count} entri ({percentage:.2f}%)")

# 2. Pengecekan Khusus untuk 'poutcome' yang menggunakan 'nonexistent'
# Ini adalah missing value implisit terbesar
poutcome_nonexistent_count = df[df['poutcome'] == 'nonexistent'].shape[0]
poutcome_nonexistent_percentage = (poutcome_nonexistent_count / total_rows) * 100

print(f"\n'poutcome' (nonexistent): {poutcome_nonexistent_count} entri ({poutcome_nonexistent_percentage:.2f}%)")

print("\nâœ… Analisis 'unknown' dan 'nonexistent' selesai.")
````

### **2.4.5 Analisis Kategorikal & Volume**

**A. Analisis Tingkat Konversi (Rate)**

````python
# 1. Definisikan kolom kategorikal kunci
cat_cols_for_analysis = ['job', 'marital', 'month', 'poutcome']

# 2. Fungsi yang ditingkatkan untuk menghitung dan memvisualisasikan Tingkat Konversi
def plot_conversion_rate_by_category(df, col):
    # Menghitung Conversion Rate: (Count 'yes') / (Total Count) * 100%
    conversion_rate = df.groupby(col)['y'].value_counts(normalize=True).mul(100).unstack().fillna(0)

    # Fokus pada persentase 'yes' dan urutkan
    conversion_rate_yes = conversion_rate['yes'].sort_values(ascending=False)

    plt.figure(figsize=(10, 6))
    ax = sns.barplot(x=conversion_rate_yes.index, y=conversion_rate_yes.values, palette='coolwarm')

    # --- MENAMBAHKAN LABEL PERSENTASE PADA SETIAP BATANG ---
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.1f}%',
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha = 'center', va = 'center',
                    xytext = (0, 9),
                    textcoords = 'offset points',
                    fontsize=9)
    # --------------------------------------------------------

    # Pengaturan plot
    plt.xticks(rotation=45, ha='right')
    plt.title(f'2.4.5 Konversi Rate (%) berdasarkan {col} (Dengan Label Persentase)', fontsize=14)
    plt.ylabel('Conversion Rate (%)')
    plt.xlabel(col)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# 3. Jalankan fungsi visualisasi
for col in cat_cols_for_analysis:
    plot_conversion_rate_by_category(df, col)
````

**ðŸ’¡ Penjelasan**

Teknik *sampling* digunakan untuk mengatasi ketidakseimbangan kelas dalam dataset (di mana Kelas 0 jauh lebih banyak daripada Kelas 1), yang dapat menyebabkan model bias. Proses ini terbagi menjadi tiga strategi utama:

1. **Undersampling:** Strategi ini berfokus pada pengurangan jumlah sampel di kelas mayoritas (Kelas 0) agar proporsinya menjadi lebih seimbang. Meskipun sederhana, teknik ini berisiko menghilangkan informasi penting yang terkandung dalam data mayoritas.

2. **Over Sampling:** Metode ini berfokus pada peningkatan jumlah sampel di kelas minoritas (Kelas 1). Ini dilakukan baik dengan menduplikasi sampel yang sudah ada (Random Oversampling) atau dengan menciptakan data baru yang disintesis (seperti SMOTE), yang merupakan teknik yang paling sering Anda gunakan untuk modeling final.

3. **Hybrid:** Kategori ini menggabungkan kedua metode di atas, seringkali menggunakan Oversampling untuk menciptakan sampel minoritas baru, dan diikuti oleh teknik undersampling ringan (Tomek Links atau ENN) untuk membersihkan batas-batas keputusan (boundary) antara kedua kelas. Teknik ini bertujuan untuk menghasilkan dataset yang paling bersih dan optimal.

## **6.2.2 SMOTE (Synthetic Minority Oversampling Technique) âœ…**

SMOTE atau Synthetic Minority Oversampling Technique adalah teknik oversampling, tetapi SMOTE bekerja secara berbeda dari oversampling pada umumnya. Pada teknik oversampling klasik, data minoritas diduplikasi dari populasi data minoritas. Teknik ini meningkatkan jumlah data, tetapi tidak memberikan informasi atau variasi baru pada model machine learning.
Oleh karena itu, Nitesh Chawla, dkk. (2002) memperkenalkan teknik baru untuk membuat data sintetis untuk tujuan oversampling dalam paper SMOTE mereka.
SMOTE bekerja dengan memanfaatkan algoritma k-nearest neighbor untuk membuat data sintetis. Pertama-tama SMOTE memilih data acak dari kelas minoritas, kemudian menetapkan k-nearest neighbors dari data tersebut. Data sintetis kemudian dibuat di antara data acak dan tetangga terdekat yang dipilih secara acak. Perhatikan contohnya di bawah ini.

smote.avif

Prosedur ini diulang beberapa kali hingga kelas minoritas memiliki proporsi yang sama dengan kelas mayoritas.

````python
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='minority')

x_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
````



Proses analisis data ini dimulai dengan membagi dataset asli (`X` dan `y`) menjadi data pelatihan (**`X_train`** dan **`y_train`**) dan data pengujian (**`X_test`** dan **`y_test`**) menggunakan fungsi **`train_test_split`** dari Scikit-learn, di mana 20% data dialokasikan untuk pengujian (`test_size=0.2`) dan sisanya 80% untuk pelatihan, sambil memastikan reproduksibilitas pemecahan dengan **`random_state=42`**. Setelah pemisahan, data pelatihan (`X_train` dan `y_train`) yang mengalami masalah **ketidakseimbangan kelas** kemudian ditangani dengan teknik *over-sampling* **SMOTE** (*Synthetic Minority Over-sampling Technique*) dari pustaka `imblearn`. Selanjutnya object **`SMOTE(sampling_strategy='minority')`** dibuat dan diterapkan menggunakan **`fit_resample`** untuk menghasilkan sampel data sintetis baru bagi kelas minoritas, menghasilkan *dataset* pelatihan yang seimbang (**`x_train_smote`** dan **`y_train_smote`**) yang siap digunakan untuk melatih model klasifikasi.

````python
print(pd.Series(y_train_smote).value_counts())
````

Dapat terlihat bahwa teknik oversampling smote telah berhasil diterapkan pada data train untuk model. Hal ini dibuktikan dengan output kode diatas, dimana jumlah data label `0` sama dengan jumlah data label `1`.

````python
from sklearn.preprocessing import StandardScaler
import pandas as pd


sc = StandardScaler()

x_train_smote = sc.fit_transform(x_train_smote)
x_test_smote = sc.transform(X_test)
````

Standarisasi Fitur merupakan proses penting untuk memastikan semua fitur (kolom) memiliki skala yang seragam. Dalam data ini dilakukan standarisasi fitur menggunakan StandardScaler dari Scikit-learn untuk mengubah semua fitur agar memiliki rata-rata nol ($\mu=0$) dan deviasi standar satu ($\sigma=1$). Pertama, sebuah objek StandardScaler diinisialisasi, kemudian diterapkan menggunakan fit_transform pada data pelatihan yang sudah di-SMOTE (x_train_smote), yang mana proses fit ini mempelajari nilai $\mu$ dan $\sigma$ dari data pelatihan tersebut. Terakhir, nilai $\mu$ dan $\sigma$ yang sama ini kemudian hanya diaplikasikan (transform) pada data pengujian (X_test) melalui sc.transform(X_test), untuk memastikan data pengujian diskalakan dengan parameter yang sama dan mencegah terjadinya kebocoran data (data leakage).

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, LinearSVC, NuSVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import recall_score, accuracy_score, roc_curve, auc, precision_score, f1_score
````

````python
names = [
    "Logistic Regression",
    "K-Nearest Neighbors",
    "Linear SVC",
    "SVC (RBF)",
    "NuSVC",
    "Gaussian Naive Bayes",
    "Random Forest",
    "AdaBoost",
    "Gradient Boosting",
    "DecisionTreeClassifier"
]

classifiers = [
    LogisticRegression(max_iter=1000),
    KNeighborsClassifier(),
    LinearSVC(max_iter=10000),
    SVC(kernel='rbf', probability=True, random_state=42),
    NuSVC(probability=True),
    GaussianNB(),
    RandomForestClassifier(),
    AdaBoostClassifier(),
    GradientBoostingClassifier(),
    DecisionTreeClassifier()
]

for name, clf in zip(names, classifiers):
    print(f"Training {name}...")
    clf.fit(x_train_smote, y_train_smote)
````

# **[Tahap 5 : Evaluation ðŸ†ðŸ†ðŸ†]**
> *Tahap 5 untuk menilai dan memvalidasi model yang sudah dilatih terhadap kriteria sukses bisnis.*

````python
results = []

for name, clf in zip(names, classifiers):

    y_pred = clf.predict(x_test_smote)

    accuracy = accuracy_score(y_test, y_pred)

    precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)

    precision_0 = precision_score(y_test, y_pred, average='binary', pos_label=0, zero_division=0)

    recall_0 = recall_score(y_test, y_pred, average='binary', pos_label=0, zero_division=0)

    f1_0 = f1_score(y_test, y_pred, average='binary', pos_label=0, zero_division=0)

    precision_1 = precision_score(y_test, y_pred, average='binary', pos_label=1, zero_division=0)

    recall_1 = recall_score(y_test, y_pred, average='binary', pos_label=1, zero_division=0)

    f1_1 = f1_score(y_test, y_pred, average='binary', pos_label=1, zero_division=0)

    results.append({
        "Model": name,
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1-Score": f1,
        "Precision (Class 0)": precision_0,
        "Recall (Class 0)": recall_0,
        "F1-Score (Class 0)": f1_0,
        "Precision (Class 1)": precision_1,
        "Recall (Class 1)": recall_1,
        "F1-Score (Class 1)": f1_1
    })


df_results = pd.DataFrame(results)

df_results['Accuracy'] = df_results['Accuracy'].map('{:.4f}'.format)
df_results['Precision'] = df_results['Precision'].map('{:.4f}'.format)
df_results['Recall'] = df_results['Recall'].map('{:.4f}'.format)
df_results['F1-Score'] = df_results['F1-Score'].map('{:.4f}'.format)
df_results['Precision (Class 0)'] = df_results['Precision (Class 0)'].map('{:.4f}'.format)
df_results['Recall (Class 0)'] = df_results['Recall (Class 0)'].map('{:.4f}'.format)
df_results['F1-Score (Class 0)'] = df_results['F1-Score (Class 0)'].map('{:.4f}'.format)
df_results['Precision (Class 1)'] = df_results['Precision (Class 1)'].map('{:.4f}'.format)
df_results['Recall (Class 1)'] = df_results['Recall (Class 1)'].map('{:.4f}'.format)
df_results['F1-Score (Class 1)'] = df_results['F1-Score (Class 1)'].map('{:.4f}'.format)

print("ðŸ“Š Hasil Evaluasi Model Klasifikasi")
df_results
````

### Hyperparameter Tunning

````python
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('model', RandomForestClassifier())
])


parameter_grid = {
    'model__max_depth': [None, 10, 20, 30],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [1, 2, 4],
    'model__n_estimators': [100, 200, 300]
}




random_search = RandomizedSearchCV(
    pipeline,
    parameter_grid,
    n_iter=500,
    cv=5,
    scoring='f1_weighted',
    n_jobs=-1,
    verbose=1
)

print("Memulai proses Randomized Search...")
random_search.fit(x_train_smote, y_train_smote)
print("Grid Search selesai.")

print("\n--- Hasil Terbaik ---")
print(f"Skor Akurasi Terbaik: {random_search.best_score_:.4f}")
print("Kombinasi Parameter Terbaik Ditemukan:")
print(random_search.best_params_)

model_terbaik = random_search.best_estimator_
````

**Cara kerja**

* Menentukan rentang nilai: Menentukan rentang atau distribusi nilai yang mungkin untuk setiap hyperparameter.
* Pengambilan sampel acak: Randomized search akan mengambil sejumlah kombinasi hyperparameter secara acak dari rentang yang telah ditentukan.
* Validasi silang: Setiap kombinasi yang diambil secara acak akan dievaluasi menggunakan validasi silang (cross-validation) untuk mengukur performanya.
* Evaluasi: Setelah semua kombinasi acak dievaluasi, randomized search akan memilih kombinasi yang menghasilkan performa terbaik.
